```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Area Government Area Summary

```python
# Calculate the total number of unique schools
school_count = school_data.count()["school_name"]
school_count
```

```python
# Calculate the total number of students
student_count = student_data.count()["student_name"]
student_count
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) maths score
average_maths_score = student_data["maths_score"].mean()
average_maths_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = student_data["reading_score"].mean()
average_reading_score
```

```python
# The percentage of passing maths (math scores greather than or equal to 50)
passing_maths_count = school_data_complete[(school_data_complete["maths_score"] >= 50)].count()["student_name"]
#passing_maths_count
passing_maths_percentage = passing_maths_count / float(student_count) * 100
passing_maths_percentage
```

```python
# The percentage of students who passed reading
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 50)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# The percentage of students that passed maths and reading
passing_maths_reading_count = school_data_complete[
    (school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)
].count()["student_name"]
overall_passing_rate = passing_maths_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Convert to DataFrame
district_summary = pd.DataFrame({"Total Schools": [school_count],
                                "Total Students": [student_count],
                                "Total Budget": [total_budget],
                                "Average Maths Score": [average_maths_score],
                                "Average Reading Score": [average_reading_score],
                                "% passing_maths_percentage": [passing_maths_percentage],
                                "% Passing Reading": [passing_reading_percentage],
                                "% Overall Passing": [overall_passing_rate]
    
})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

# School Summary

```python
# Use the code provided to select all of the school types
school_types = school_data.sort_values("school_name")["type"]
school_types
```

```python
# Calculate the total student count per school
# per_school_counts = school_data[['school_name', 'size']]
# per_school_counts 

grp_by_school = school_data_complete.groupby('school_name')
per_school_counts = grp_by_school.count()["student_name"]
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = school_data.sort_values("school_name")['budget']
per_school_capita = per_school_budget/list(per_school_counts)
per_school_budget
per_school_capita
```

```python
# Calculate the average test scores per school
per_school_maths = grp_by_school.mean('maths_score')["maths_score"]
per_school_reading = grp_by_school.mean('reading_score')["reading_score"]
per_school_maths
per_school_reading
```

```python
# Calculate the number of students per school with maths scores of 50 or higher
students_passing_maths = school_data_complete[
    (school_data_complete["maths_score"] >= 50)
]
school_students_passing_maths = students_passing_maths.groupby(["school_name"]).size()
school_students_passing_maths
```

```python
# Calculate the number of students per school with reading scores of 50 or higher
students_passing_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 50)]
school_students_passing_reading = students_passing_reading.groupby(["school_name"]).size()
school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both maths and reading with scores of 50 or higher
students_passing_maths_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 50) & (school_data_complete["maths_score"] >= 50)
]
school_students_passing_maths_and_reading = students_passing_maths_and_reading.groupby(["school_name"]).size()
school_students_passing_maths_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_maths = school_students_passing_maths / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_maths_and_reading / per_school_counts * 100
overall_passing_rate
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({
                                'School Type':list(school_types),
                                'Total Students':list(per_school_counts),
                                'Total School Budget':list(per_school_budget),
                                'Per Student Budget':list(per_school_capita),
                                'Average Maths Score':per_school_maths,
                                'Average Reading Score':per_school_reading,
                                '% Passing Maths':per_school_passing_maths,
                                '% Passing Reading':per_school_passing_reading,
                                '% Overall Passing':overall_passing_rate
    
})

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

# Top Performing Schools (By % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.
top_schools = per_school_summary.sort_values("% Overall Passing", ascending= False)
top_schools.head(5)
```

# Bottom Performing Schools (By % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.
bottom_schools = per_school_summary.sort_values("% Overall Passing")
bottom_schools.head(5)
```

# Maths Scores by Year

```python
# Create data series of scores by Year levels using conditionals
year_nine = school_data_complete[(school_data_complete["year"] == 9)]
year_ten = school_data_complete[(school_data_complete["year"] == 10)]
year_eleven = school_data_complete[(school_data_complete["year"] == 11)]
year_twelve = school_data_complete[(school_data_complete["year"] == 12)]

# Group each by school name
year_nine_maths_scores = year_nine.groupby("school_name")
year_ten_maths_scores = year_ten.groupby("school_name")
year_eleven_maths_scores = year_eleven.groupby("school_name")
year_twelve_maths_scores = year_twelve.groupby("school_name")

# Combine series into single DataFrame
maths_scores_by_Year = pd.DataFrame({
    "year 9": year_nine_maths_scores["maths_score"].mean(),
    "year 10": year_ten_maths_scores["maths_score"].mean(),
    "year 11": year_eleven_maths_scores["maths_score"].mean(),
    "year 12": year_twelve_maths_scores["maths_score"].mean()})


# Minor data wrangling
maths_scores_by_Year.index.name = None 

# Display the DataFrame
maths_scores_by_Year
```

# Reading Score by Year

```python
# Create data series of scores by Year levels using conditionals
year_nine = school_data_complete[(school_data_complete["year"] == 9)]
year_ten = school_data_complete[(school_data_complete["year"] == 10)]
year_eleven = school_data_complete[(school_data_complete["year"] == 11)]
year_twelve = school_data_complete[(school_data_complete["year"] == 12)]

# Group each by school name
year_nine_reading_scores = year_nine.groupby("school_name")
year_ten_reading_scores = year_ten.groupby("school_name")
year_eleven_reading_scores = year_eleven.groupby("school_name")
year_twelve_reading_scores = year_twelve.groupby("school_name")

# Combine series into single DataFrame
reading_scores_by_year = pd.DataFrame({
    "year 9": year_nine_reading_scores["reading_score"].mean(),
    "year 10": year_ten_reading_scores["reading_score"].mean(),
    "year 11": year_eleven_reading_scores["reading_score"].mean(),
    "year 12": year_twelve_reading_scores["reading_score"].mean()
})


# Minor data wrangling
reading_scores_by_year.index.name = None

# Display the DataFrame
reading_scores_by_year
```

# Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()

#school_spending_df.dtypes
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].map(lambda x: float(x.lstrip('$')))
school_spending_df
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['Per Student Budget'], spending_bins, labels=labels, include_lowest=True)
school_spending_df
```

```python
#  Calculate averages for the desired columns.
spending_maths_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_maths = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Maths"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    'Average Maths Score': spending_maths_scores,	
    'Average Reading Score': spending_reading_scores,
    '% Passing Maths': spending_passing_maths,	
    '% Passing Reading': spending_passing_reading,	
    '% Overall Passing': overall_passing_spending
}) 

# Display results
spending_summary
```

# Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
per_school_summary["School Size"] = pd.cut(per_school_summary['Total Students'], size_bins, labels=group_names)
per_school_summary
```

```python
size_maths_scores = per_school_summary.groupby(["School Size"])["Average Maths Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_maths = per_school_summary.groupby(["School Size"])["% Passing Maths"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
size_summary = pd.DataFrame({'Average Maths Score': size_maths_scores,
                            'Average Reading Score': size_reading_scores,
                            '% Passing Maths': size_passing_maths,
                            '% Passing Reading': size_passing_reading,
                            '% Overall Passing': size_overall_passing
})

# Display results
size_summary
```
